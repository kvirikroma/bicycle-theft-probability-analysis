```python
import os

import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import utils, Sequential

from repository.insurance_premium_estimation_repository import (_simple_insurance_premium_prediction, generate_random_insurance_data,
                                                                LockType, FrameMaterial, BikeType, WK_DEVICE_VERSIONS,
                                                                MIN_PRICE, MAX_PRICE, MIN_LOCK_PRICE, MAX_LOCK_PRICE, MAX_SECONDS_IN_MONTH)

SAVES_PATH_RELATIVE = 'checkpoints/'
SAVES_PATH = os.path.join(os.path.dirname(SAVES_PATH_RELATIVE), SAVES_PATH_RELATIVE)
BATCH_SIZE = 256

data_x, data_y = [*zip(*(([
    i.bike_price,
    i.lock_type.value,
    i.bike_type.value,
    i.frame_material.value,
    i.parking_time_during_last_month,
    i.user_risk_tendency.avg_theft_probability_prediction,
    i.user_risk_tendency.avg_theft_probability,
    i.user_risk_tendency.avg_recovery_probability_prediction,
    i.user_risk_tendency.avg_recovery_probability,
    i.user_risk_tendency.avg_parking_time_theft_probability_prediction,
    i.user_risk_tendency.avg_parking_time,
    i.lock_price,
    i.wk_device_revision_number,
    i.bike_is_electric,
    i.damage_insurance_included,
], _simple_insurance_premium_prediction(i)) for i in (generate_random_insurance_data() for _ in range(1000000))))]
print(data_x[:5])
print(data_y[:5])
```

```python
data_x_prepared = []
max_parking_time = max(i[9] for i in data_x)

def normalize(x, minimum, maximum):
    return (x - minimum) / (maximum - minimum)

for item in data_x:
    data_x_prepared.append([
        normalize(item[0], MIN_PRICE, MAX_PRICE),
        *[i.value == item[1] for i in LockType],
        *[i.value == item[2] for i in BikeType],
        *[i.value == item[3] for i in FrameMaterial],
        normalize(item[4], 0.0, MAX_SECONDS_IN_MONTH),
        *item[5:10],
        normalize(item[10], 0, MAX_SECONDS_IN_MONTH) if item[10] < MAX_SECONDS_IN_MONTH else 1.0,
        normalize(item[11], MIN_LOCK_PRICE, MAX_LOCK_PRICE),
        *[i == item[12] for i in WK_DEVICE_VERSIONS],
        *item[13:]
    ])

x_size = len(data_x_prepared[0])
data_x_prepared = np.array(data_x_prepared)
data_y_prepared = np.array([0.0 if i is None else i for i in data_y])

print(data_x_prepared[:5])
print(data_y_prepared[:5])
print()
print(data_x_prepared.shape)
print(data_y_prepared.shape)
```

```python
model = Sequential([
    Input((x_size, )),
    Dense(x_size * 4, activation="relu"),
    Dense(x_size * 8, activation="relu"),
    Dense(x_size * 8, activation="relu"),
    Dense(x_size * 4, activation="relu"),
    Dense(1, activation="linear")
])
model.compile(tf.keras.optimizers.Adam(), loss=tf.keras.losses.MeanSquaredError())
```

```python
model.summary()
utils.plot_model(model, show_shapes=True)
```

```python
history = model.fit(
    data_x_prepared,
    data_y_prepared,
    epochs=10,
    batch_size=BATCH_SIZE,
    shuffle=True,
    # callbacks=[
        # tf.keras.callbacks.ModelCheckpoint(
        #     SAVES_PATH + "checkpoint-{epoch:02d}-{val_loss:.6f}.hdf5",
        #     monitor='val_loss',
        #     save_weights_only=True
        # ),
        # tf.keras.callbacks.EarlyStopping(
        #     monitor='val_loss', min_delta=0.000001, patience=2, verbose=1
        # )
    # ],
    verbose=True,
    validation_split=0.2
)
# model.save_weights(SAVES_PATH + "result.hdf5")
```
